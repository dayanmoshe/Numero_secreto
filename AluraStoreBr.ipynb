### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja = loja['Preço'].sum() + loja['Frete'].sum()
faturamento_loja_semfrete = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum() + loja2['Frete'].sum()
faturamento_loja2_semfrete = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum() + loja3['Frete'].sum()
faturamento_loja3_semfrete = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum() + loja4['Frete'].sum()
faturamento_loja4_semfrete = loja4['Preço'].sum()

print('Faturamento da Loja 1:' + str(faturamento_loja))
print('Faturamento da Loja 1 sem frete:' + str(faturamento_loja_semfrete))
print('Faturamento da Loja 2:' + str(faturamento_loja2))
print('Faturamento da Loja 2 sem frete:' + str(faturamento_loja2_semfrete))
print('Faturamento da Loja 3:' + str(faturamento_loja3))
print('Faturamento da Loja 3 sem frete:' + str(faturamento_loja3_semfrete))
print('Faturamento da Loja 4:' + str(faturamento_loja4))
print('Faturamento da Loja 4 sem frete:' + str(faturamento_loja4_semfrete))
```

```python
import matplotlib.pyplot as plt

fig, fat_4lojas = plt.subplots()

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos_com_frete = [faturamento_loja, faturamento_loja2, faturamento_loja3, faturamento_loja4]
faturamentos_sem_frete = [faturamento_loja_semfrete, faturamento_loja2_semfrete, faturamento_loja3_semfrete, faturamento_loja4_semfrete]


fat_4lojas.bar(lojas, faturamentos_com_frete, label='Com frete')
bar_labels = ['Loja 1', 'Loja 2', 'Loja 3', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

fat_4lojas.bar(lojas, faturamentos_com_frete, label=bar_labels, color=bar_colors)

fat_4lojas.set_ylabel('valor faturado com frete')
fat_4lojas.set_title('Fatturamento das lojas do Sr Joao COM FRETE')

plt.show()
```
